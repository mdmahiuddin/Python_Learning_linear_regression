```python
#we are going to learn linear regression model with only one variable 
#to do the linear regression, we need to bring some library

import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
#need to bring the data file

df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
#change the variable name to make the next steps easy, otherwise we need to adopt different method. It will have in another file

df.rename(columns={'per capita income (US$)':'PerCapitaIncome'}, inplace=True)
```

```python
#we can show the information with simple graph
%matplotlib inline
plt.xlabel('year')
plt.ylabel('PerCapitaIncome')
plt.scatter(df.year, df.PerCapitaIncome, color ='red', marker='+')
```

```python
#drop the per capita income column
new_df = df.drop('PerCapitaIncome', axis = 'columns')
new_df
```

```python
#see only per capita income column
PerCapitaIncome = df.PerCapitaIncome
PerCapitaIncome
```

```python
#creating linear regression object using two separate column (year and per capita income)
reg = linear_model.LinearRegression()
reg.fit(new_df, PerCapitaIncome)
```

```python
#let us predict the per capita income for 2020

reg.predict([[2020]])
```

```python
#here is the regression coefficient

reg.coef_
```

```python
#use csv file to predict per capita predictions
year_df =  pd.read_csv('YearForPerCapitaIncomePrediction.csv')
year_df
```

```python

#all predicted value
p = reg.predict(year_df)
p
```

```python
#see the year and their predicted value
year_df['PerCapitaIncome']=p
year_df
```

```python
#Let us save it 
year_df.to_csv('IncomePrediction.csv')
```
